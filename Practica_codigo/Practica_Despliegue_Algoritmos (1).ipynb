```python
%%writefile funciones.py

import pandas as pd
import numpy as np
import mlflow
import time
import mlflow.sklearn
import subprocess
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_breast_cancer
import argparse
from itertools import product
from pyngrok import ngrok

##Arguments
def argumentos():
    parser = argparse.ArgumentParser(description='Argumentos de entrada')
    parser.add_argument('--nombre_run', type=str, help='Nombre para el entrenamiento')
    parser.add_argument('--n_estimators_list', nargs='+', type=int, help='Lista de n_estimators')
    parser.add_argument('--max_depth_list', nargs='+', type=int, help='Lista de max_depth')
    parser.add_argument('--learning_rate_list', nargs='+', type=float, help='Lista de learning_rate')
    return parser.parse_args()

def load_dataset():
    cancer = load_breast_cancer()
    df = pd.DataFrame(cancer["data"], columns=cancer["feature_names"])
    df["target"] = cancer["target"]
    return df

# Train pipeline
def data_treatment(df):
    train, test = train_test_split(df, test_size=0.2)
    test_target = test["target"]
    features = [x for x in list(train.columns) if x != 'target']
    x_raw = train[features]
    y_raw = train['target']
    x_train, x_test, y_train, y_test = train_test_split(x_raw, y_raw,
                                                        test_size=0.2,
                                                        random_state=1,
                                                        stratify=y_raw)
    return x_train, x_test, y_train, y_test

def mlflow_tracking(n_estimators, learning_rates, max_depths, nombre_run, x_train, x_test, y_train, y_test):
    # Kill any existing ngrok tunnels
    ngrok.kill()

    # Start MLflow server on port 5000
    mlflow_ui_process = subprocess.Popen(['mlflow', 'ui', '--port', '5000'])

    # Allow some time for the server to start
    time.sleep(5)

    # Authenticate with Ngrok and create tunnel
    NGROK_AUTH_TOKEN = '2gQ4sJufZCkp3C97yTaF7jcyyQb_4QDEjvNJupZzvyLBbYcGW'
    ngrok.set_auth_token(NGROK_AUTH_TOKEN)
    ngrok_tunnel = ngrok.connect(5000, "http")
    print('El tracking UI:', ngrok_tunnel.public_url)

    # Set MLflow tracking URI to the Ngrok tunnel
    mlflow.set_tracking_uri(ngrok_tunnel.public_url)

    mlflow.set_experiment('Pruebas desde script')
    # Iterate over all possible combinations
    for n_estimator, learning_rate, max_depth in product(n_estimators, learning_rates, max_depths):
        with mlflow.start_run(run_name=nombre_run) as run:
            bt = GradientBoostingClassifier(random_state=42,
                                            max_depth=max_depth,
                                            learning_rate=learning_rate,
                                            n_estimators=n_estimator)

            preprocessor = Pipeline(steps=[('scaler', StandardScaler())])
            model = Pipeline(steps=[('preprocessor', preprocessor), ('GradientBoostingClassifier', bt)])
            model.fit(x_train, y_train)
            accuracy_train = model.score(x_train, y_train)
            accuracy_test = model.score(x_test, y_test)

            mlflow.log_metric('accuracy_train', accuracy_train)
            mlflow.log_metric('accuracy_test', accuracy_test)
            mlflow.log_param('n_estimators', n_estimator)
            mlflow.log_param('learning_rate', learning_rate)
            mlflow.log_param('max_depth', max_depth)
            mlflow.sklearn.log_model(model, 'clf-model')
```

```python
import funciones as func
```

```python
%%writefile main.py

import funciones as func

def main():
  print('Iniciando entrenamiento del modelo')
  args_values = func.argumentos()
  df = func.load_dataset()
  x_train, x_test, y_train, y_test = func.data_treatment(df)
  func.mlflow_tracking(args_values.n_estimators_list, args_values.learning_rate_list, args_values.max_depth_list, args_values.nombre_run, x_train, x_test, y_train, y_test)

if __name__ == '__main__':
  main()
```
